```python
# split the data to train and validation
X_train, X_valid, y_train, y_valid = train_test_split(X_resampled, y_resampled, test_size=0.15, random_state=42, shuffle=True)
```

```python
# show the categorical and numerical features from the data

categorical_features = [col for col in X_resampled.columns if X_resampled[col].dtype not in["float32", "float64", "int32", "int64"]]
numercal_features = [col for col in X_resampled.columns if X_resampled[col].dtype in["float32", "float64", "int32", "int64"]]
```

```python
# Pipeline for data preprocessing

def pipeline_for_processing(numerical_cols, categorical_cols):
    
    if numerical_cols and categorical_cols:

        # numerical pipeline
        num_pipeline = Pipeline(steps=[
                                ("selector", DataFrameSelector(numerical_cols)),
                                ("imputer", SimpleImputer(strategy="median")),
                                ("scaler", StandardScaler())])

        # categorical pipeline
        cate_pipeline = Pipeline(steps=[
                                ("selector", DataFrameSelector(categorical_cols)),
                                ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
                                ("encode", OneHotEncoder(sparse_output=False)) ])


        # total pipeline
        total_pipeline = FeatureUnion(transformer_list=[
                                ("num_pipeline", num_pipeline),
                                ("cate_pipeline", cate_pipeline)])
        return total_pipeline
    
    elif numerical_cols :
        
        # numerical pipeline
        num_pipeline = Pipeline(steps=[
                                ("selector", DataFrameSelector(numerical_cols)),
                                ("imputer", SimpleImputer(strategy="median")),
                                ("scaler", StandardScaler())])
        return num_pipeline
    
    elif categorical_cols:
        
        # categorical pipeline
        cate_pipeline = Pipeline(steps=[
                                ("selector", DataFrameSelector(categorical_cols)),
                                ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
                                ("encode", OneHotEncoder(sparse_output=False)) ])
        
        return cate_pipeline
    else:
        raise ValueError("Both numerical_cols and categorical_cols are empty or None.")

pipline_func = pipeline_for_processing(numercal_features, categorical_features)
```

```python
# fit the total pipeline
pipline_func.fit(X_train)
X_train_preprocessed = pipline_func.transform(X_train)
X_valid_preprocessed = pipline_func.transform(X_valid)
```

Model And Hyperparameter tuning

```python
RF_clf = RandomForestClassifier()

params = {
    "n_estimators" : np.arange(100, 200, 50),
    "max_depth" : np.arange(5, 20, 5),
    "min_samples_split" : np.arange(2, 7, 2),
    "min_samples_leaf" : np.arange(1, 7, 2)
    }

RandomSearch = GridSearchCV(estimator=RF_clf, param_grid=params, scoring="accuracy", n_jobs=-1, cv=5, verbose=1)
RandomSearch.fit(X_train, y_train)
```

```python
# get the best hyperparameters
RandomSearch.best_params_
```

```python
RF_best_clf =  RandomSearch.best_estimator_
RF_best_clf
```

```python
cross_validation = cross_val_score(estimator=RF_best_clf, X=X_train, y=y_train, scoring="accuracy", n_jobs=-1, cv=5, verbose=1)
print(f"The scores from cross validation score {cross_validation}")
print("--"*50)
print(f"The mean of the scores {cross_validation.mean()}")
print("**"*50)

# make cross validation to predict
cross_validation_pred = cross_val_predict(estimator=RF_best_clf, X=X_train, y=y_train, method="predict",
                                         cv=5, n_jobs=-1, verbose=1)

# Compute the confusion matrix
confusion = confusion_matrix(y_train, cross_validation_pred)

# Create a heatmap with Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', cbar=False,
            annot_kws={'size': 14, 'weight': 'bold'}, linewidths=.5)

# Customize labels and title
plt.xlabel('Predicted', fontsize=14, fontweight='bold')
plt.ylabel('Actual', fontsize=14, fontweight='bold')
plt.title('Confusion Matrix', fontsize=16, fontweight='bold')
plt.show()
```
